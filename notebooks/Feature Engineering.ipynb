```python
# match-level dataframe
import pandas as pd
import numpy as np

MASTER_CSV = r"C:\Users\A_R COMPUTERS\Desktop\Vineet\Project\IPL Prediction\ipl_master.csv"
master = pd.read_csv(MASTER_CSV)

# One row per match
matches = master.drop_duplicates("match_id").copy()
matches = matches.sort_values("match_date").reset_index(drop=True)

print(matches.shape)
```

```python
# Did toss winner win the match?
matches["toss_winner_won"] = (matches["toss_winner"] == matches["winner"]).astype(int)

print(matches["toss_winner_won"].value_counts())
```

```python
matches["toss_decision_encoded"] = matches["toss_decision"].map({"bat": 0, "field": 1})

print(matches["toss_decision_encoded"].value_counts())
```

```python
matches
```

```python
# Total wins per team (all time)
total_wins = matches.groupby("winner")["match_id"].count().reset_index()
total_wins.columns = ["team", "total_wins"]

print(total_wins.sort_values("total_wins", ascending=False))
```

```python
matches
```

```python
#  Total matches played per team
team1_count = matches.groupby("team1")["match_id"].count().reset_index()
team1_count.columns = ["team", "matches_as_team1"]

team2_count = matches.groupby("team2")["match_id"].count().reset_index()
team2_count.columns = ["team", "matches_as_team2"]

total_matches = team1_count.merge(team2_count, on="team")
total_matches["total_matches"] = total_matches["matches_as_team1"] + total_matches["matches_as_team2"]

print(total_matches[["team", "total_matches"]].sort_values("total_matches", ascending=False))
```

```python
matches
```

```python
# Overall win rate per team
team_stats = total_matches.merge(total_wins, on="team")
team_stats["win_rate"] = round(team_stats["total_wins"] / team_stats["total_matches"], 2)

print(team_stats[["team", "total_matches", "total_wins", "win_rate"]].sort_values("win_rate", ascending=False))
```

```python
# Attach win rate to each match for team1 and team2
matches = matches.merge(team_stats[["team", "win_rate"]], left_on="team1", right_on="team", how="left")
matches = matches.rename(columns={"win_rate": "team1_win_rate"}).drop(columns="team")

matches = matches.merge(team_stats[["team", "win_rate"]], left_on="team2", right_on="team", how="left")
matches = matches.rename(columns={"win_rate": "team2_win_rate"}).drop(columns="team")

matches["win_rate_diff"] = matches["team1_win_rate"] - matches["team2_win_rate"]

print(matches[["team1", "team2", "team1_win_rate", "team2_win_rate", "win_rate_diff"]].head(10))
```

```python
# Head to head wins
h2h = matches.groupby(["team1", "winner"])["match_id"].count().reset_index()
h2h.columns = ["team1", "winner", "h2h_wins"]

print(h2h.head(10))
```

```python
h2h
```

```python
# Venue win rate per team
# How many times each team won at each venue
venue_wins = matches.groupby(["venue", "winner"])["match_id"].count().reset_index()
venue_wins.columns = ["venue", "team", "venue_wins"]

# How many times each team played at each venue
venue_played_t1 = matches.groupby(["venue", "team1"])["match_id"].count().reset_index()
venue_played_t1.columns = ["venue", "team", "venue_played"]

venue_played_t2 = matches.groupby(["venue", "team2"])["match_id"].count().reset_index()
venue_played_t2.columns = ["venue", "team", "venue_played"]

venue_played = pd.concat([venue_played_t1, venue_played_t2]).groupby(["venue", "team"])["venue_played"].sum().reset_index()

# Merge and compute win rate
venue_stats = venue_played.merge(venue_wins, on=["venue", "team"], how="left")
venue_stats["venue_wins"] = venue_stats["venue_wins"].fillna(0)
venue_stats["venue_win_rate"] = round(venue_stats["venue_wins"] / venue_stats["venue_played"], 2)

print(venue_stats.sort_values("venue_win_rate", ascending=False).head(10))
```

```python
# Target variable 
matches["target"] = (matches["winner"] == matches["team1"]).astype(int)

print("Target distribution:")
print(matches["target"].value_counts())
print("Team1 win %:", round(matches["target"].mean() * 100, 2), "%")
```
